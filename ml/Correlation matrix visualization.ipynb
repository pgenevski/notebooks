```python
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns


def make_triu_mask(matrix):
    mask = np.zeros_like(matrix)
    mask[np.triu_indices_from(mask)] = True
    return mask

corr = np.random.random((4,4)) - 0.5
ax = plt.axes((1,1,2,2))
_ = sns.heatmap(corr,
            annot=True,
            mask=make_triu_mask(corr),
            square=True,
            cmap="RdYlGn", # https://matplotlib.org/tutorials/colors/colormaps.html  # bwr, RdYlGn
            xticklabels=['inner','water','cool','fuel'],
            yticklabels=['inner','water','cool','fuel'],
            ax=ax)
```
