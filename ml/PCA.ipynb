```python
%matplotlib inline

import numpy as np
import pandas as pd

from sklearn import datasets
from sklearn.decomposition import PCA

import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

import seaborn as sns
sns.set(color_codes=True)


iris = datasets.load_iris()
df = pd.DataFrame(iris.data,columns=iris.feature_names)
sns.pairplot(df)
```

```python
cov = np.cov(iris.data,rowvar=False)
sns.heatmap(cov,annot=True, xticklabels=iris.feature_names,yticklabels=iris.feature_names)
```

```python
corr = np.corrcoef(iris.data,rowvar=False)
sns.heatmap(corr,annot=True,xticklabels=iris.feature_names,yticklabels=iris.feature_names)
```

```python
pca = PCA(n_components=2,whiten=False).fit(iris.data)
print('components:\n',pca.components_.T)
print('explained ratio:\n',pca.explained_variance_ratio_)

names = iris.target_names[iris.target]
projected = np.matmul(iris.data,pca.components_.T) 

df = pd.DataFrame(projected,columns=['pc1','pc2'])
sns.scatterplot(data=df,x='pc1',y='pc2',hue=names)
```

```python
reduced = pca.transform(iris.data)

df = pd.DataFrame(reduced,columns=['pc1','pc2'])
sns.scatterplot(data=df,x='pc1',y='pc2',hue=names)
```
